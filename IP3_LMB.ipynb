5.	Remove all the symbols (\$\,)from the 'Price' column and convert  and update it to  64-bit int numbers using astype(‘int64’)

Hint. Look for string functions like replace/strip).

```python
car_sales['Price'] = car_sales['Price'].str.replace('$', '').str.replace(',', '').astype('int64')
car_sales
```

6.	Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function.

```python
df = pd.DataFrame(car_sales)
df['Total Sales'] = df['Price'].cumsum()
print(df)
```

7.	Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today). Check out “date_range” function and keep the periods=len(car_sales). 

```python
start_date = datetime.today().strftime('%Y-%m-%d')
df['Sale Date'] = pd.date_range(start=start_date, periods=len(car_sales))
print(df)
```

8.	Use the matplotlib line plot function to plot the 'Price' column versus the 'Total Sales' column (Give proper title and appropriate labels.)

```python
Price = df['Price']
Total_Sales = df['Total Sales']
plt.plot(Price, Total_Sales)
plt.title('Line Plot of Price vs Total Sales')
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.show()
```

9.	Create a matplotlib scatter plot of the 'Odometer (KM)' vs 'Price' column.

```python
Odometer = df['Odometer (KM)']
plt.scatter(Odometer, Price)
plt.title('Scatter Plot of Odometer (KM) vs Price')
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.show()
```

12.	Create a subplot of multiple plots ( step 8,9,10,11).

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# First plot
sns.lineplot(x='Price', y='Total Sales', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Line Plot of Price vs Total Sales')



# Second plot
sns.scatterplot(x='Odometer (KM)', y='Price', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Scatter Plot of Odometer (KM) vs Price')


# Third plot
sns.barplot(x='Make', y='Odometer (KM)', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Bar Graph of Make vs Odometer (KM)')


# Fourth plot
sns.histplot(df['Make'], kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Histogram of Make')
```

16.	Find out Pearson correlation between 'Total Sales','Price','Odometer (KM)'. Comment on the correlation 

```python
columns_of_interest = ['Price', 'Odometer (KM)', 'Total Sales']
correlation_matrix = df[columns_of_interest].corr(method='pearson')
plt.figure(figsize=(5, 4))  # Adjust the size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, vmin=-1, vmax=1)
plt.show()
```

A negative Pearson correlation of -0.85 between Price and Odometer (KM) suggests a strong inverse relationship. This means that as the odometer reading increases (more kilometers driven), the car's price tends to decrease.
